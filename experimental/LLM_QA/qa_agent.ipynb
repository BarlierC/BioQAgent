## LLM-based QA Agent (RAG)

### Aim

The goal of this section was to extend the retrieval system into a full **Retrieval-Augmented Generation (RAG) pipeline** by integrating a **Large Language Model (LLM)** to generate natural language answers grounded in retrieved PubMed abstracts.

### Attempted Approaches

Two main approaches were considered or explored:
1. OpenAI LLM
2. Local LLM inference using Ollama (with models such as `llama3` and `phi3`)

### Observed Limitations

1. **OpenAI** approach:
- Requires an **API key and billing setup**
- Not compatible with the constraint of a fully free and reproducible project

2. **Local LLM (Ollama)**: while conceptually appealing (fully local, no cost), this approach had significant practical limitations:
- **High CPU usage** during inference
- Long execution time (no response after several minutes)
- **System instability** (kernel crashes, overheating)

These issues are primarily due to:
- The computational cost of transformer-based models (even small ones)
- The large input context (multiple biomedical abstracts)
- Lack of GPU acceleration on a standard personal laptop

### Outcome

Due to these constraints, the **LLM-based QA pipeline could not be reliably executed end-to-end**.

As a result:
- No stable outputs could be generated
- Quantitative or qualitative evaluation of generated answers was not possible
